```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.datasets import fetch_20newsgroups
import random
import tkinter
from tkinter import *
from tkinter import Text
import tkinter.messagebox
r = tkinter.Tk()
r.geometry("200x200")
r["bg"]="gray"
def click():
        root =tkinter.Tk()
        categories = ['alt.atheism','comp.graphics','sci.med','soc.religion.christian']
        twenty_train = fetch_20newsgroups(subset='train',categories=categories, shuffle=True, random_state=42)       
        #root.grid()
        v = StringVar()
        v.set("1")
        def sel():   
          src_category=v.get()        
          for (idx,val) in enumerate(twenty_train.target_names):
            if (val==src_category):
                src_category_index=idx
          docs_of_src_category=[]          #this list contains the index of all documents of the category user entered
          for (idx,val) in enumerate(twenty_train.data):        
            if(twenty_train.target[idx]==src_category_index):
                docs_of_src_category.extend([idx])        
          random_index=random.randrange(0,len(docs_of_src_category))
          src_index=docs_of_src_category[random_index]
          #resultStr.set( "A random document is selected from the category %s as the source document"+str(src_index))
         # var = StringVar()
         # label = Message( root, textvariable=var, relief=RAISED )
          #var.set("Hey!? How are you doing?%s %s"%(src_category,twenty_train.data[src_index]))
         # label.grid()        
          tkinter.messagebox.showinfo("Hello!","\nA random document is selected from the category %s as the source document\n%s"%(src_category,twenty_train.data[src_index]))
        Radiobutton(root,text = "Altheism", variable=v, 
        value="alt.atheism",command=sel).grid(row = 0, column = 0, sticky = W)
        Radiobutton(root, text = "Christianity", variable=v, 
        value="soc.religion.christian",command=sel).grid(row = 1, column = 0, sticky = W)
        Radiobutton(root, text = "Computer Graphics", variable=v, 
        value="comp.graphics",command=sel).grid(row = 2, column = 0, sticky = W)
        Radiobutton(root, text = "Medical Science", variable=v, 
        value="sci.med",command=sel).grid(row = 3, column = 0, sticky = W) 
        root.mainloop()
def inputWindow():
      top = tkinter.Tk()
      top.geometry("750x450+150+50")
      top["bg"]="gray"
      
      def saveCallback():
          text1 =t.get('1.0', 'end-1c')          
          with open("example1.txt", 'w') as outfile:
                 outfile.write(text1)           
      def saveCallback1():
           text2 = t1.get('1.0', 'end-1c')
           with open("example2.txt", 'w') as outfile:
                outfile.write(text2)
         
      def cmp(): 
          s=open('C:\Users\Sifat\Documents\Python Scripts\Plagiarism Detection\example1.txt')
          s2=open('C:\Users\Sifat\Documents\Python Scripts\Plagiarism Detection\example2.txt')
          
          data_set=(s.read(),s2.read()) 
         # m = SequenceMatcher(None,s.read(),s2.read())
          count_vect = CountVectorizer()
          term_freq_matrix=count_vect.fit_transform(data_set)
          tfidf = TfidfTransformer()
          tf_idf_matrix = tfidf.fit_transform(term_freq_matrix)
          similarity=cosine_similarity(tf_idf_matrix[0], tf_idf_matrix[1])
          r=similarity[0][0]*100         
         # a=m.ratio()*100
          tkinter.messagebox.showinfo("RESULT","Similarity between the documents is %f!"%r)                  
      L1=Label(top,text="INPUT",fg="blue")
      L1.grid(row=0,column=3)
      L2=Label(top,text=" ")
      L2.grid(row=2,column=1)
      t =Text(top,height=20,width=40)
      t.grid(row=2,column=2)
      t1=Text(top,height=20,width=40)
      t1.grid(row=2,column=4)
      b1 = tkinter.Button(top, text ="Save",fg='blue',command = saveCallback) 
      b1.grid(row=3,column=2)   
      b3=tkinter.Button(top,text="Compare",fg="blue",command=cmp)
      b3.grid(row=4,column=3)
      b4=tkinter.Button(top,text="Save",fg="blue",command=saveCallback1)
      b4.grid(row=3,column=4)           
      top.mainloop()
B=tkinter.Button(r,text="Custominput Window",command=inputWindow)
B.pack()
E=tkinter.Button(r,text="Evaluation using 20NewsGroup",command=click)
E.pack()
r.mainloop()
```
